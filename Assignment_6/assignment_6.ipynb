```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weather.csv')
df.head()
```

```python
print(df.columns)
print(df.shape)
df.describe(include = 'all')
```

```python
# Visualization 1 - Seeing relation between MaxTemp and Evaporation, if high temp results in more Evaporation
sns.jointplot(x='MaxTemp', y='Evaporation', data=df)
```

```python
#Visualization 2 - To check relation between temperature and pressure at 3pm
sns.jointplot(data= df,kind="hex",x='Temp3pm',y="Pressure3pm")
```

```python
#Visualization 3 - Seeing if rainfall follows normal distribution or not
sns.distplot(df['Rainfall'], kde = True)
```

```python
#Visualization 4 Histogram of Mintemp
sns.distplot(df['MinTemp'], kde=False, rug=True)
```

```python
#Visualization 5 - Check if Cloud9am follows Normal distribution

sns.kdeplot(df['Cloud9am'], shade=True)
```

```python
#Visualization 6

sns.kdeplot(df['RISK_MM'], label='original')
sns.kdeplot(df['RISK_MM'], bw=0.2, label='bw=0.2')
sns.kdeplot(df['RISK_MM'], bw=2, label='bw=2')
plt.legend()
```

```python
#Visualization 7 : checking if clouds effect humidity in the region at 3pm

sns.jointplot(x= df['Cloud3pm'], y=df['Humidity3pm'], data=df, kind='kde')
```

```python
#Visualization 8 Relation bw Humidity and Temperature at 9 am

df2=df.loc[:, ['Humidity9am','Temp9am']]
sns.pairplot(df2)
```

```python
#Visualization 9 : How likely Rain today effects Sunshine
sns.stripplot(x = 'RainToday', y='Sunshine', data=df)
```

```python
#Visualization 10 If clouds present at 9 am bring rain 

sns.swarmplot(x='RainToday', y='Cloud9am', data=df)
```

```python
#Visualization 11: Relation bw Min temp and Rain

sns.boxplot(x='RainToday', y='MinTemp', data=df,hue='RainTomorrow', dodge=True)
```

```python
#Visualization 12
sns.boxplot(x='RainToday', y='MaxTemp', data=df)
```

```python
#Visualization 13
sns.barplot(x='Location', y='WindGustSpeed', hue='RainToday', data=df)
```

```python
#Visualization 14 Most popular Wind Gust Direction
sns.countplot(x='WindGustDir', data=df, palette='Greens_d')
```

```python
# Visualization 15 Relation between Humidity and rainfall
df.plot(kind='scatter', x = 'Humidity9am', y = 'Rainfall')
```

```python
#Visualization 16
df2=df.loc[:, ['Humidity3pm','Cloud3pm']]
sns.pairplot(df2)
```

```python
#Visualization 17
sns.countplot(x='RainToday', data=df, palette='rocket')
```

```python
#Visualization 18
sns.swarmplot(x='RainTomorrow', y='Cloud9am', data=df)
```

```python
#Visualization 19
sns.scatterplot(data=df,x="Evaporation",y='Humidity3pm')
```

```python
#Visualization 20
sns.countplot(x='WindDir9am', data=df, palette='rocket')
```
